```python
#Data visualization using Matplotlib

import pandas as pd

df=pd.read_csv(r'D:\BigData_All_content\csv\Fruit_data.csv')
df.head()
```

```python
#importing matplotlib pkg

import matplotlib.pyplot as plt
```

```python
#Creating line chart for the yield of 'Apple'
plt.plot(df.Year, df.Apple)
```

```python
#Giving XLabel, YLabel and Title of the line chart

plt.plot(df.Year, df.Apple)
plt.title("Yield of Apple in Lucknow")
plt.xlabel('Year')
plt.ylabel("Yield for Apple (Tons per hectare)")
```

```python
#Line chart for Orange
plt.plot(df.Year, df.Orange)
plt.title("Yield of Orange in Lucknow")
plt.xlabel('Year')
plt.ylabel("Yield for Orange (Tons per hectare)")
```

```python
#Creating line chart for 3 lines

plt.plot(df.Year, df.Apple)
plt.plot(df.Year, df.Orange)

plt.title('Fruit yield in Lucknow')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
```

```python
#We can add a legend which tells us what each line in out graph means

plt.plot(df.Year, df.Apple)
plt.plot(df.Year, df.Orange)

plt.title('Fruit yield in Lucknow')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.legend(['Apple', 'Orange'])
```

```python
#We can add marker info in a line chart. We may use different marker symbols:
#circle (o), cross (x), square (s) and diamond (d)

plt.plot(df.Year, df.Apple, marker='s')
plt.plot(df.Year, df.Orange, marker='o')

plt.title('Fruit yield in Lucknow')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.legend(['Apple', 'Orange'])
```

```python
plt.plot(df.Year, df.Apple, marker='d')
plt.plot(df.Year, df.Orange, marker='x')

plt.title('Fruit yield in Lucknow')
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
plt.legend(['Apple', 'Orange'])

#We can save any graph using savefig()
plt.savefig("e:\\fruit_data.jpg", dpi=600)
```

```python
df2=pd.read_csv(r'D:\BigData_All_content\csv\Admission_Predict.csv')
df2.head()
```

```python
#Creating scatter plot on 'GRE Score' and 'CGPA'

plt.scatter(df2['GRE Score'], df2['CGPA'])
plt.xlabel('GRE Score')
plt.ylabel('CGPA')
plt.title('GRE Score V/S CGPA')
```

```python
df2.describe()
```

```python
#Observation

#From the above graph, it is clear that candidate with higher CGPA usually have higher GRE Score

#There is one candidate having higher GRE Score but having lower CGPA
```

```python
#Creating scatter plot on 'TOEFL Score' and 'CGPA'

plt.scatter(df2['TOEFL Score'], df2['CGPA'])
plt.xlabel('TOEFL Score')
plt.ylabel('CGPA')
plt.title('TOEFL Score V/S CGPA')
```

```python
#Ex: Observing the relation ship b/w GRE Score and CGPA of those students whose CGPA=8.0

x=df2[df2['CGPA']>=8.0]
x.head(10)
```

```python
plt.scatter(x['GRE Score'], x['CGPA'], color='red')
plt.xlabel('GRE Score')
plt.ylabel('CGPA>=8.0')
plt.title('GRE Score V/S CGPA when CGPA>=8.0')
```

```python
#Creating histogram on 'GRE Score'
df2['GRE Score'].plot(kind='hist', color='green', bins=100)
plt.xlabel('GRE Score')
plt.ylabel('Frequency')
plt.title('Frequency of Data in GRE Score')
```

```python
#Observation

#There is density b/w 310 and 330 for GRE Score
#GRE Score>330, would be a good score for a candidate to get admission
```

```python
#Creating histogram on 'TOEFL Score'
df2['TOEFL Score'].plot(kind='hist', color='green', bins=100)
plt.xlabel('TOEFL Score')
plt.ylabel('Frequency')
plt.title('Frequency of Data in TOEFL Score')
```

```python
#Bar chart

#To show the frequency of each category in 'Research' using bar chart

fr=df2.Research.value_counts()
fr
```

```python
fr.plot(kind='bar', color=['pink', 'cyan'])
plt.xlabel('Research')
plt.ylabel('Frequency')
plt.title("Bar chart for Research")
```

```python
uni_count=df2['University Rating'].value_counts()
uni_count
```

```python
uni_count.plot(kind='bar', color=['red', 'green', 'yellow', 'pink', 'cyan'])
```

```python
#Usually bar chart is created to represent categorical data

#Comparing TOEFL Score graphically

best=df2['TOEFL Score'].max()
worst=df2['TOEFL Score'].min()
avg=df2['TOEFL Score'].mean()

import numpy as np

y1=np.array([best, avg, worst])
x1=['Best', 'Average', 'Worst']

plt.bar(x1, y1, color=['Green', 'Yellow', 'red'])
plt.xlabel('Level')
plt.ylabel('TOEFL Score')
plt.title("Comparison of TOEFL Score")
```

```python
#Pie chart
sizes = [10, 20, 45, 25]
labels = ["Cats", "Dogs", "Tigers", "Goats"]
plt.pie(sizes, labels = labels)
plt.axes().set_aspect("equal")
plt.show()
```
